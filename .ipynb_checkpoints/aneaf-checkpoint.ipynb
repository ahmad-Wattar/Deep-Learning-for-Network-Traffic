# Einleitung

Im Rahmen der Studienleistung **„Maschinelles Lernen und IT-Sicherheit“** wird in diesem Projekt Netzwerkverkehr mithilfe von Methoden des maschinellen Lernens analysiert und klassifiziert. Grundlage der Untersuchung sind Netzwerkmitschnitte, die mit **Wireshark** unter verschiedenen Nutzungsszenarien (z. B. Webzugriffe, Dateiübertragungen) aufgezeichnet wurden. Die einzelnen Mitschnitte wurden anschließend gemischt und in Form von **Flow-Daten** zur weiteren Analyse bereitgestellt.

Ziel des Projekts ist es, zunächst mithilfe von **Unsupervised-Learning-Verfahren** Strukturen und Muster im Netzwerkverkehr zu identifizieren, ohne auf vorab bekannte Labels zurückzugreifen. Die so entstehenden Cluster werden analysiert und gelabelt. Auf Basis dieser erzeugten Labels wird anschließend ein **Deep-Learning-Modell** trainiert, das Netzwerkverkehr selbstständig klassifizieren kann.

Das Projekt ist in zwei Hauptteile gegliedert:

- **Unsupervised Learning** zur Clusterbildung und Labelgenerierung  
- **Deep Learning** zur überwachten Klassifikation des Netzwerkverkehrs  

Alle Schritte, von der Datenvorbereitung über die Modellierung bis hin zur Evaluation werden vollständig in diesem Jupyter Notebook dokumentiert und nachvollziehbar dargestellt.

# Expolarative Datenanalyse

```python
# Einlesen der Excel-Tabelle mit pandas und Konvertierung in eine CSV-Datei
df = pd.read_excel('aneaf.xlsx')
df.to_csv('aneaf.csv', index=False)
print("Konvertierung abgeschlossen")
```

- **Src_IP**: Quell-IP-Adresse des Netzwerkverkehrs  
- **Dst_IP**: Ziel-IP-Adresse des Netzwerkverkehrs  
- **Src_Port**: Quell-Port der Verbindung  
- **Dst_Port**: Ziel-Port der Verbindung  
- **Protocol**: Verwendetes Protokoll (z. B. TCP, UDP, DNS)  
- **Duration**: Dauer des Flows in Sekunden  
- **Total_Pkts**: Gesamtanzahl der Pakete im Flow  
- **Total_Bytes**: Gesamtanzahl der übertragenen Bytes  
- **Pkt_Len_Std**: Standardabweichung der Paketlängen  
- **Pkt_Len_Max**: Maximale Paketlänge  
- **Payload_Bytes_Total**: Summe der Nutzdaten-Bytes (ohne Header)  
- **IAT_Std**: Standardabweichung der Inter-Arrival-Times (Zeit zwischen Paketen), ein kleiner Wert deutet auf einen regelmäßigen Traffic  
- **IAT_Max**: Maximale Inter-Arrival-Time zwischen zwei Paketen  


```python
# Die ersten 5 Flows anzeigen
data = pd.read_csv("aneaf.csv")
data.head(5)
```

```python
# Protokoll-Verteilung
print(data['Protocol'].value_counts())

# Port-Verteilung 
print("\nTop 10 Destination Ports:")
print(data['Dst_Port'].value_counts().head(10))
```

### Die Analyse der Protokollverteilung zeigt eine klare Dominanz von TCP-basiertem Verkehr, ergänzt durch einen signifikanten Anteil an DNS-Flows. Darüber hinaus treten UDP-Verbindungen sowie vereinzelte ICMPv6-Pakete auf, die potenziell als Ausreißer oder Hintergrundverkehr betrachtet werden können.



```python
eps = 1e-6 # Keine Division durch null
data["Payload_Ratio"] = data["Payload_Bytes_Total"] / (data["Total_Bytes"] + eps) #Wie viel ist Nutzdaten vs. Header/Overhead?
data["Avg_Pkt_Size"]  = data["Total_Bytes"] / (data["Total_Pkts"] + eps) #Durchschnittliche Paketgröße
data["Byte_Rate"]     = data["Total_Bytes"] / (data["Duration"] + eps) #Bytes per Second
data["Pkt_Rate"]      = data["Total_Pkts"] / (data["Duration"] + eps) #Packets per Second
data["Avg_Time_per_Pkt"] = data["Duration"] / (data["Total_Pkts"] + eps) #Durchschnittliche Zeit per Paket
```

```python
data.head(5)
```

- **Payload_Ratio**: Anteil der Nutzdaten an allen Bytes (Payload / Total_Bytes)  
- **Avg_Pkt_Size**: Durchschnittliche Paketgröße (Total_Bytes / Total_Pkts)  
- **Pkt_Rate**: Pakete pro Sekunde (Total_Pkts / Duration)  
- **Byte_Rate**: Bytes pro Sekunde (Total_Bytes / Duration)
- **Avg_Time_per_Pkt**: Durchschnittliche Zeit zwischen zwei Paketen, also wie lange ein Paket im Schnitt braucht.

## Einiges logarithmieren und dann skalieren 

```python
data = data.copy()


log_features = [
    "Total_Bytes",
    "Total_Pkts",
    "Payload_Bytes_Total",
    "Byte_Rate",
    "Pkt_Rate"
]

no_log_features = [
    "Duration",
    "Pkt_Len_Std",
    "Pkt_Len_Max",
    "Avg_Pkt_Size",
    "IAT_Std",
    "IAT_Max",
    "Avg_Time_per_Pkt",
    "Payload_Ratio"
]

# Log-Transformation
for f in log_features:
    data[f] = np.log1p(data[f])

X = data[log_features + no_log_features]

# Skalierung
X_scaled = StandardScaler().fit_transform(X)
```

```python
X_scaled
```

## Anzeigen nach der Skalierung 

```python
X_scaled_df = pd.DataFrame(
    X_scaled,
    columns=X.columns,
    index=X.index
)
```

```python
for col in X_scaled_df.columns:
    plt.figure(figsize=(5,3))
    plt.hist(X_scaled_df[col], bins=50)
    plt.title(f"Skaliert (log falls angewendet): {col}")
    plt.xlabel("Skalierter Wert")
    plt.ylabel("Häufigkeit")
    plt.tight_layout()
    plt.show()
```

```python
X_scaled_df
```

# Schritt 2 

## PCA Anwenden 

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print("Erklärte Varianz:", pca.explained_variance_ratio_)
print("Kumulierte Varianz:", pca.explained_variance_ratio_.sum())
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], s=8, alpha=0.6)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA-Projektion der Flow-Daten (ohne Clustering)")
plt.tight_layout()
plt.savefig("pca.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Elbow anwenden um die Anzahl der Cluster festzustellen 

```python
inertias = []
k_range = range(2, 11)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertias.append(km.inertia_)
```

```python
plt.figure(figsize=(6,4))
plt.plot(k_range, inertias, marker="o")
plt.xlabel("Anzahl Cluster (k)")
plt.ylabel("Inertia")
plt.title("Elbow-Methode für K-Means")
plt.tight_layout()
plt.savefig("Elbow.png", dpi=300, bbox_inches="tight")

plt.show()
```

# 6 Clusters

```python
k = 6
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)

labels = kmeans.fit_predict(X_scaled)
```

```python
unique, counts = np.unique(labels, return_counts=True)
dict(zip(unique, counts))
```

```python
data["cluster"] = labels
```

```python
plt.figure(figsize=(6,5))
plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=labels,
    cmap="tab10",
    s=10,
    alpha=0.7
)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA-Projektion mit K-Means-Clustern (k=6)")
plt.colorbar(label="Cluster")
plt.tight_layout()
plt.savefig("k_means4.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
cluster_protocol = pd.crosstab(
    data["cluster"],
    data["Protocol"],
    normalize="index"   # Prozent pro Cluster
)

cluster_protocol
```

### DBSCAN für einen Vergleich
#### Mit dem **k-Distance-Plot** wird ein sinnvoller Wert für **eps** bestimmt. Dabei wurde **k = 15** festgelegt, d. h. für jeden Punkt wird die **Distanz zu seinem 15.-nächsten Nachbarn** berechnet. Diese Distanzen werden sortiert geplottet; der **Knick** der Kurve dient als Orientierung für **eps**.


```python
k = 15  # gleich min_samples
nbrs = NearestNeighbors(n_neighbors=k).fit(X_scaled)
distances, _ = nbrs.kneighbors(X_scaled)

distances = np.sort(distances[:, k-1])

plt.figure(figsize=(6,4))
plt.plot(distances)
plt.xlabel("Datenpunkte (sortiert)")
plt.ylabel(f"{k}-NN-Distanz")
plt.title("k-Distance-Plot zur eps-Wahl")
plt.tight_layout()
plt.savefig("eps.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
pca = PCA(n_components=2, random_state=42)
X_pca2 = pca.fit_transform(X_scaled)
```

```python
dbscan = DBSCAN(eps=1.7, min_samples=15)
db_labels = dbscan.fit_predict(X_scaled)

plt.figure(figsize=(6,5))
plt.scatter(
    X_pca2[:, 0],
    X_pca2[:, 1],
    c=db_labels,
    cmap="tab10",
    s=8,
    alpha=0.7
)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA-Projektion mit DBSCAN-Clustern")
plt.colorbar(label="Cluster (-1 = Noise)")
plt.tight_layout()
plt.show()
```

```python
dict(zip(*np.unique(db_labels, return_counts=True)))
```

```python
silhouette_score(X_scaled, db_labels)
```

###  Begründung für die Wahl von *k = 5* (k-means) und reduzierten Features

Die Wahl von **k = 5** basiert auf einer kombinierten datengetriebenen Analyse unter Verwendung aller Features sowie eines reduzierten Feature-Sets.

####  Silhouette Scores – alle Features
- **k = 6:** 0.56  
- **k = 5:** 0.55  
- **k = 4:** 0.52  
- **k = 3:** 0.54  

####  Silhouette Scores – reduzierte Features
- **k = 6:** 0.56  
- **k = 5:** 0.55  
- **k = 4:** 0.54  
- **k = 3:** 0.54  

Zusätzlich identifizierte DBSCAN eine natürliche Clusterstruktur mit **fünf stabilen Clustern** (zzgl. Noise), was die Existenz von fünf dominanten Dichtezentren im Merkmalsraum nahelegt.

Obwohl der höchste Silhouette Score bei *k = 6* erreicht wird, liegt der Wert für *k = 5* in beiden Feature-Sets nur marginal darunter und zeigt eine **vergleichbare Clusterqualität**.  
Gleichzeitig bietet *k = 5* eine höhere **semantische Interpretierbarkeit** und reduziert die Fragmentierung großer Cluster, was insbesondere für die Analyse von Netzwerkdaten vorteilhaft ist.



####  Begründung für die Verwendung reduzierter Features
Die reduzierten Features wurden gewählt, da die ursprünglichen Merkmale eine **starke Korrelation** untereinander aufweisen und somit eine hohe Redundanz enthalten.  
Durch die Feature-Reduktion wird:

- die **Modellkomplexität** verringert  
- die **Rechenzeit** beim Training signifikant reduziert  
- die **Effizienz der Clusterbildung** erhöht  
- das Risiko von **verzerrten Distanzmaßen** durch korrelierte Dimensionen minimiert  
- die **Stabilität der Clusterstruktur** verbessert

Die folgende Tabelle zeigt die relative Verteilung der Netzwerkprotokolle DNS, TCP und UDP innerhalb der fünf Cluster:

####  Interpretation der Clusterstruktur

Die Protokollverteilung zeigt eine **klare semantische Differenzierung** der Cluster:

- **Cluster 0:** DNS-dominierter Traffic  
  → typisch für Namensauflösung, Service Discovery, Hintergrundkommunikation

- **Cluster 1:** Gemischter Traffic (TCP/UDP)  
  → gemischte Anwendungen wie Streaming

- **Cluster 2:** Stark TCP-dominierter Traffic  
  → typische Client-Server-Kommunikation (z.B. Web, Downloads, Services)

- **Cluster 3:** TCP-dominierter Spezialverkehr  
  → besondere TCP-Verbindungen mit anderem Nutzungsverhalten

- **Cluster 4:** TCP-lastiger Verkehr mit UDP-Anteil  
  → überwiegend TCP, mit kleineren UDP-Anteilen



##### Obwohl die Cluster 2, 3 und 4 eine ähnliche Protokollverteilung aufweisen (starke TCP-Dominanz), werden sie durch k-means aufgrund unterschiedlicher statistischer Kommunikationsmerkmale getrennt. Die Cluster unterscheiden sich somit nicht auf Protokollebene, sondern auf struktureller Traffic-Ebene (z.B. Sessioncharakteristik, Volumen, Zeitverhalten, Paketstatistik).



Diese Verteilung bestätigt, dass die Clusterbildung nicht zufällig erfolgt, sondern eine **inhaltlich interpretierbare Segmentierung des Netzwerkverkehrs** widerspiegelt.



**Fazit:**  
Die Kombination aus datengetriebener Clusteranalyse (DBSCAN, Silhouette Score), Feature-Reduktion und der beobachteten Protokollstruktur bestätigt **k = 5** als eine fachlich, statistisch und semantisch valide Wahl für die Segmentierung der Netzwerkdaten.


## Hierarisches Clustering mit dem  linkage="average" für alle Features

```python
X = X_scaled_df.values

metrics = ["euclidean", "manhattan", "cosine"]
k_range = range(2, 10)

results = []


for m in metrics:
    for k in k_range:
        try:
            hc = AgglomerativeClustering(
                n_clusters=k,
                linkage="average",
                metric=m
            )
            labels = hc.fit_predict(X)
            sil = silhouette_score(X, labels, metric=m)
            results.append({
                "metric": m,
                "k": k,
                "silhouette": sil
            })
            print(f"{m} | k={k} | silhouette={sil:.4f}")
        except Exception as e:
            print(f"{m} | k={k} | ERROR:", e)

results_df = pd.DataFrame(results)
```

```python
pivot = results_df.pivot(index="metric", columns="k", values="silhouette")

plt.figure(figsize=(10,4))
sns.heatmap(pivot, annot=True, fmt=".2f", cmap="viridis")
plt.title("Silhouette Scores – Hierarchical Clustering (Average-Linkage)")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Metric")
plt.show()
```

## Modellauswahl

### Modellwahl: Multilayer Perceptron (MLP)

Für die Klassifikation des Netzwerkverkehrs wurde ein **Multilayer Perceptron (MLP)** als neuronales Netz gewählt.  
Diese Modellwahl basiert auf der Struktur der vorliegenden Daten.

Die verwendeten Netzwerkdaten liegen in **tabellarischer Form** vor und bestehen aus extrahierten numerischen Merkmalen (Flows, aggregierte Statistikfeatures, Kommunikationsmetriken).  
Es handelt sich dabei nicht um zeitlich sequenzielle Rohdaten (wie Paketfolgen) und nicht um räumlich strukturierte Daten (wie Bilder).

Daher sind spezialisierte Architekturen wie **Convolutional Neural Networks (CNNs)** oder **Recurrent Neural Networks (LSTMs)** nicht geeignet:

- **CNNs** sind für räumliche Mustererkennung optimiert (z.B. Bilder, Matrixstrukturen) und benötigen lokale Korrelationen in Gitternetzstrukturen, die in tabellarischen Feature-Vektoren nicht vorhanden sind.
- **LSTMs** sind für zeitliche Abhängigkeiten und Sequenzmodellierung konzipiert (z.B. Zeitreihen, Sprachdaten, Paketfolgen), während die vorliegenden Daten bereits aggregierte Merkmalsrepräsentationen darstellen und keine expliziten zeitlichen Sequenzen enthalten.

Ein **MLP** ist hingegen speziell für die Verarbeitung **hochdimensionaler, numerischer Feature-Vektoren** geeignet und kann nichtlineare Zusammenhänge zwischen den Merkmalen effizient modellieren.  
Durch die Verwendung mehrerer versteckter Schichten (Hidden Layers) kann das Modell komplexe Entscheidungsgrenzen im Merkmalsraum lernen, ohne unnötige strukturelle Annahmen über räumliche oder zeitliche Abhängigkeiten zu treffen.

---

Für das neuronale Netz wird zunächst eine grundlegende Architektur auf Basis von Fully Connected Layers (Dense Layern) definiert.

**Dense Layer** werden eingesetzt, da sie nichtlineare Kombinationen der Eingabefeatures lernen können und besonders für tabellarische, numerische Datensätze geeignet sind.

Als Aktivierungsfunktion wird **ReLU (Rectified Linear Unit)** verwendet, da sie eine effiziente Gradientenweitergabe ermöglicht und sich als Standardaktivierungsfunktion in Klassifikationsaufgaben etabliert hat.

Für die Ausgabeschicht wird eine **Softmax-Aktivierung** verwendet, da es sich um eine Multiclass-Klassifikationsaufgabe handelt.  
Die Softmax-Funktion erzeugt eine Wahrscheinlichkeitsverteilung über die Klassen und ermöglicht eine eindeutige Klassenzuordnung.

Die Output-Schicht besteht aus **5 Neuronen**, entsprechend der fünf definierten Verkehrsklassen.

Die konkrete Anzahl der Neuronen in den Hidden Layers wird im nächsten Schritt experimentell bestimmt und durch Evaluationsmetriken (Accuracy, Precision, Recall, F1-Score) optimiert.



## Training des Modells 

```python
X = dataset.drop(columns=["label", "label_name"], errors="ignore").values
y = dataset["label"].astype(int).values
num_classes = len(np.unique(y))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=41, stratify=y
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)

# Für reproduzierbare Ergebnisse
tf.random.set_seed(42)
np.random.seed(42)

#Modell
def build_model(input_dim, num_classes, n_hidden=1, neurons=32, activation="relu"):
    model = Sequential()
    model.add(Input(shape=(input_dim,)))

    for _ in range(n_hidden):
        model.add(Dense(neurons, activation=activation))

    model.add(Dense(num_classes, activation="softmax"))

    model.compile(
        optimizer="adam",
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model


# mögliche Kombinitionen

hidden_layers_list = [1, 2, 3]
neurons_list = [16, 32, 64]
activations = ["relu", "tanh", "sigmoid"]

EPOCHS = 15
BATCH_SIZE = 32

results = []


#Schleife über Kombinationen

for n_hidden in hidden_layers_list:
    for neurons in neurons_list:
        for act in activations:
            tf.keras.backend.clear_session()

            model = build_model(
                input_dim=X_train.shape[1],
                num_classes=num_classes,
                n_hidden=n_hidden,
                neurons=neurons,
                activation=act
            )

            hist = model.fit(
                X_train, y_train,
                validation_split=0.2,
                epochs=EPOCHS,
                batch_size=BATCH_SIZE,
                verbose=0
            )

            test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
            best_val_acc = max(hist.history["val_accuracy"])

            results.append({
                "hidden_layers": n_hidden,
                "neurons": neurons,
                "activation": act,
                "best_val_acc": best_val_acc,
                "test_acc": test_acc
            })

results_df = pd.DataFrame(results).sort_values(by="test_acc", ascending=False)
results_df
```

```python
# für reproduzierbare Ergebnisse
tf.random.set_seed(42)
np.random.seed(42)

# EarlyStopping
early = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

# Eins der besten Modelle  aus der Tabelle: 2 Hidden, 16 Neuronen, relu
tf.keras.backend.clear_session()
model = build_model(
    input_dim=X_train.shape[1],
    num_classes=num_classes,
    n_hidden=2,
    neurons=64,
    activation="relu"
)

hist = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=100,          # hohe Obergrenze, EarlyStopping stoppt früher
    batch_size=32,
    callbacks=[early],
    verbose=1
)

# Test-Evaluation
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Test loss: {test_loss:.4f}")
print(f"Test acc : {test_acc:.4f}")

# Confusion Matrix 
y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))
```

```python
# Confusion Matrix berechnen
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(7,6))
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()

ticks = np.arange(cm.shape[0])
plt.xticks(ticks, ticks)
plt.yticks(ticks, ticks)
plt.xlabel("Predicted label")
plt.ylabel("True label")

thresh = cm.max() / 2
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j],
                 ha="center", va="center")

plt.tight_layout()
plt.show()
```

```python
N_RUNS = 6
BASE_SEED = 42

results_runs = []  

for run in range(1, N_RUNS + 1):
    seed = BASE_SEED + run

    # Seeds setzen
    tf.keras.backend.clear_session()
    tf.random.set_seed(seed)
    np.random.seed(seed)

    # Split (jedes Mal anders)
    X = dataset.drop(columns=["label", "label_name"], errors="ignore").values
    y = dataset["label"].astype(int).values
    num_classes = len(np.unique(y))

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=seed, stratify=y
    )

    # Skalierung (fit nur auf Train)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # EarlyStopping
    early = tf.keras.callbacks.EarlyStopping(
        monitor="val_loss", patience=5, restore_best_weights=True
    )

    # Modell
    model = build_model(
        input_dim=X_train.shape[1],
        num_classes=num_classes,
        n_hidden=2,
        neurons=64,
        activation="relu"
    )

    # Training
    hist = model.fit(
        X_train, y_train,
        validation_split=0.2,
        epochs=100,
        batch_size=32,
        callbacks=[early],
        verbose=0
    )

   
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)

    #Metriken
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    macro_f1 = f1_score(y_test, y_pred, average="macro")
    weighted_f1 = f1_score(y_test, y_pred, average="weighted")

    cm = confusion_matrix(y_test, y_pred)

    print(f"\n=== RUN {run}/{N_RUNS} (seed={seed}) ===")
    print(f"Epochs: {len(hist.history['loss'])}")
    print(f"Test acc: {test_acc:.4f}")
    print(f"Balanced acc: {bal_acc:.4f} | Macro F1: {macro_f1:.4f} | Weighted F1: {weighted_f1:.4f}")
    print("Confusion Matrix:\n", cm)

    results_runs.append({
        "run": run,
        "seed": seed,
        "epochs": len(hist.history["loss"]),
        "test_acc": test_acc,
        "balanced_acc": bal_acc,
        "macro_f1": macro_f1,
        "weighted_f1": weighted_f1
    })

    # Plot Confusion Matrix
    plt.figure(figsize=(7, 6))
    plt.imshow(cm, interpolation="nearest")
    plt.title(f"Confusion Matrix - Run {run} (acc={test_acc:.4f})")
    plt.colorbar()

    ticks = np.arange(cm.shape[0])
    plt.xticks(ticks, ticks)
    plt.yticks(ticks, ticks)
    plt.xlabel("Predicted label")
    plt.ylabel("True label")

    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, cm[i, j], ha="center", va="center")

    plt.tight_layout()
    plt.show()

#Zusammenfassung über alle Runs
results_df_runs = pd.DataFrame(results_runs)
print("\n=== SUMMARY (mean over runs) ===")
print(results_df_runs[["test_acc","balanced_acc","macro_f1","weighted_f1"]].mean())
```

### Ergebnisse des Deep Learning Modells

### Eignung für reale Netzwerke
Ein wichtiger Punkt ist, dass die Labels **Pseudo-Labels aus dem Clustering** sind. Das Modell lernt damit vor allem, die im Feature-Raum gefundenen Cluster-Grenzen zu reproduzieren.  
Daher vermute ich, dass die Leistung auf **anderen Netzwerkdaten** geringer ausfallen würde. Für einen realen Einsatz wären **Ground-Truth-Labels** notwendig.
